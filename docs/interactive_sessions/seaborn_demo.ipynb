# Introduction to data visualization with seaborn

Basic visual exploration

## Attach packages and load the penguins data

```python
import seaborn as sns
import palmerpenguins # pip install palmerpenguins in Terminal if you don't have it
from palmerpenguins import load_penguins
```

```python
penguins = load_penguins()
penguins.head()
```

Wooo penguins in Python! 

## Basic data viz in seaborn

Note: Examples below are directly from https://github.com/mcnakhaee/palmerpenguins (palmerpenguins package in Python developer). Thank you! 

```python
g = sns.boxplot(x = 'island',
            y ='body_mass_g',
            hue = 'species',
            data = penguins,
            palette=['#FF8C00','#159090','#A034F0'],
            linewidth=0.3)
g.set_xlabel('Island')
g.set_ylabel('Body Mass')
```

```python
g = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=7,
               data=penguins,
               palette=['#FF8C00','#159090','#A034F0'])
g.set_xlabels('Flipper Length')
g.set_ylabels('Body Mass')
```
